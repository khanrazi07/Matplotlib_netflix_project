```python
#how many movies and tv shows
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd

data = pd.read_csv('netflix_titles.csv')

df = pd.DataFrame(data)

movies_count = (df['type'] == 'Movie').sum()
tv_shows = (df['type'] == 'TV Show').sum()

categories = ['Movie' , 'TV Show']
counts = [movies_count , tv_shows]



plt.bar(categories , counts , color = ['blue' , 'red'] )
plt.grid(color = 'gray' , linestyle = ':')
plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Counts")
plt.savefig("Movies_vs_TV_Shows.png" , dpi = 300 , bbox_inches = 'tight')
plt.tigh
plt.show()




        
        


```

```python
#percentage of each content rating
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd

data = pd.read_csv('netflix_titles.csv')

df = pd.DataFrame(data)
rating_counts = df['rating'].value_counts()
plt.figure(figsize = (12,12))
plt.pie(rating_counts , labels = rating_counts.index , autopct = "%1.1f%%" , startangle = 90)
plt.title("Content Ratings")
plt.tight_layout()
plt.savefig("Content_rating.png" , dpi = 200 )
plt.show()

```

```python
#How has the number of releases changed every year
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd

data = pd.read_csv('netflix_titles.csv')

df = pd.DataFrame(data)
release_change = (df['release_year'].value_counts().sort_index())
plt.figure(figsize = (15,15))
plt.scatter(release_change.index , release_change.values , color = 'blue' , marker = "o")
plt.grid(color = "gray" , linestyle = ":")
plt.title("Release Changes" , fontsize = 20)
plt.xlabel("Release Years" , fontsize  = 15)
plt.ylabel("Number of Shows" , fontsize = 15)
plt.tight_layout()
plt.savefig("Release_year_isight.png" , dpi = 300)
plt.show()


```

```python
#Distribution of movies duration
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd

data = pd.read_csv('netflix_titles.csv')

df = pd.DataFrame(data)
movies_df = df[df['type'] == 'Movie'].copy()
df['duration'] = df['duration'].fillna(df['duration'].mode()[0])
movies_df['duration'] = movies_df['duration'].str.replace("min" , "").astype(float)
plt.figure(figsize = (10,10))
plt.hist(movies_df['duration'] , bins = 30 ,  color = 'purple' , edgecolor = 'black')
plt.xlabel("Movies")
plt.ylabel("Duration")
plt.title("Distribution of Movies Duration")
plt.grid(color = "grey" , linestyle = ':')
plt.tight_layout()
plt.savefig("Movies_Duration_insight.png" , dpi = 300)
plt.show()
```

```python
#relationship between releases and number of shows
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd

data = pd.read_csv('netflix_titles.csv')
df = pd.DataFrame(data)
plt.figure(figsize=(12,12))
release_change = (df['release_year'].value_counts().sort_index())
plt.scatter(release_change.index , release_change.values , color = "red" )
plt.title("Release vs Number of Shows")
plt.xlabel("Shows")
plt.ylabel("Years")
plt.tight_layout()
plt.grid(color = 'gray' , linestyle = ":")
plt.show()
plt.savefig("Release_vs_Shows.png" , dpi = 300)
```

```python
#top 10 countries wiht highest number of shows
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd

data = pd.read_csv('netflix_titles.csv')
df = pd.DataFrame(data)
plt.figure(figsize=(12,12))

country_count = df['country'].value_counts().head(10)
plt.barh(country_count.index , country_count.values , color = 'red' , edgecolor = "black")
plt.title("Top 10 Countries by Number of Shows" , fontsize = 20)
plt.xlabel("Number of Shows" , fontsize = 15)
plt.ylabel('Countries' , fontsize = 15)
plt.tight_layout()
plt.savefig("Countries_Shows.png" , dpi = 300)
plt.show()



```

```python
#comparing multiple plots together
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd

data = pd.read_csv('netflix_titles.csv')
df = pd.DataFrame(data)

content_year = df.groupby(['release_year' , 'type']).size().unstack().fillna(0)

fig , ax = plt.subplots(1,2,figsize=(25,15))

ax[0].plot(content_year.index , content_year['Movie'] , color = 'blue' , linestyle = "--" , marker = "o")
ax[0].grid(color = "gray" , linestyle = ":")
ax[0].set_title("Movie Released per year", fontsize=30)
ax[0].set_xlabel("Years", fontsize=20)
ax[0].set_ylabel("Number of Movies", fontsize=20)

ax[1].plot(content_year.index , content_year['TV Show'] , color = 'red' , linestyle = "--" , marker = "o")
ax[1].grid(color = "gray" , linestyle = ":")
ax[1].set_title("TV Show Released per year", fontsize = 30)
ax[1].set_xlabel("Years" , fontsize=20)
ax[1].set_ylabel("Number of TV Show", fontsize=20)

plt.suptitle("Movies vs TV Shows" , fontsize = 40)
plt.tight_layout()

plt.savefig("Movies_vs_TV_shows_Released.png")
plt.show()


```
